```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from importlib import reload
reload(plt)
```

```python
#parameters
g = 1
L = 1
theta0 = 0.9 #in radians
pt0 = 0.9
```

```python
def F(x,t):
  return [x[1],
          -(g/L)* np.sin(x[0])
  ]
```

```python
t = np.linspace(0, 100, 100)
solution = odeint(F,[theta0 ,pt0],t)
print(solution[:5])
theta = solution[:,0]
p_theta = solution[:,1]
plt.figure(figsize=(4, 4))
plt.plot(theta, p_theta, '.')
plt.title("\u03B8={i}, p\u03B8={j}".format(i=theta0, j=pt0))
plt.xlabel("\u03B8")
plt.ylabel("$p_{\u03B8}$")
```
