```python
#https://mathworld.wolfram.com/Circle-CircleIntersection.html
import matplotlib.pyplot as plt
import numpy as np
import math

#Defining some constants
R_S = 6.957e+8                    #Solar radius (in meters) = R
r_E = 6.378e+6                    #Earth radius (in meters) = r
R = 1
r = r_E/R_S
print(r)
au = 1.496e+11                    #astronomical unit (1 A.U.) = 1.496e+11 m

#Defining some orbital parameters
a = 0.1 * au/R_S                  #semimajor axis of orbit in solar radius
i_degrees = 87.4                  #angle of orbital inclination in degrees
i = math.pi*(i_degrees/180)       #angle of orbital inclination in radians     x/180 = math.radians(x/math.pi)
y = a * math.cos(i)
F = []                            #list collecting flux

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
plt.rcParams['axes.formatter.useoffset'] = False

def D(x):
  d = np.sqrt(x**2 + y**2)
  for j in d:
    if j >= r+R:
      F.append(1)
    elif j <= R-r :
      F.append(1 - (np.pi* r**2)/(np.pi* R**2))
    else:
      A = r**2 * np.arccos((j**2 + r**2 - R**2)/(2*j*r)) + R**2 * np.arccos((j**2 + R**2 - r**2)/(2*j*R)) - (0.5)*np.sqrt((-j+r+R)*(j+r-R)*(j-r+R)*(j+r+R))
      F.append(1- A/(np.pi* R**2))
  return np.array(F)



x = np.linspace(-(r+R), (r+R), 100)
plt.title("0.1 A.U. i={k}$^\circ$".format(k=i_degrees))
plt.ylabel("Flux Ratio")
plt.xlabel("Phase")
plt.plot(x, D(x), color='blue')
plt.show()
```
