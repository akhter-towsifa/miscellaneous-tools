```python
import numpy as np
import matplotlib.pyplot as plt
from importlib import reload
reload(plt)
```

```python
#parameters
C     = 3.5 #3.5, 4, 4.5
y_dot = 0.1 #0.1, 1, 3, 5 # We have y_dot>0. So, randomly picking some positive integers or rational numbers
```

```python
x = np.linspace(-10, 10, 100000)
def H(x):
  return np.sqrt(-C-y_dot**2+ x**2 + (8*x)/(4*x**2-1))
  #this is the equation for x-dot but positive values only
def F(x):
  return -np.sqrt(-C-y_dot**2+ x**2 + (8*x)/(4*x**2-1))
  #this is the equation for x-dot but negative values only
plt.figure(figsize=(4,4))
plt.plot(x,H(x), 'b')
plt.plot(x,F(x), 'b')
plt.ylim(-10, 10)
plt.title('C={i}   at   $y\u0307$={j}'.format(i=C, j=y_dot))
plt.xlabel("x")
plt.ylabel("$p_x$")
plt.show()
```
