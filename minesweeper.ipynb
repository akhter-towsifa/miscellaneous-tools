```python
import time
import numpy as np
```

```python
class MineSweeper:
    
    def __init__(self, mode):
        mode_grid_dict = {
            'simple': 10,
            'intermediate': 25,
            'expert': 50
        }
        
        self.mode = mode_grid_dict[mode.lower()]
        print("Let's play Minesweeper! >:kapow:<")
        # time.sleep(1)

    def game_start(self):
        cell = input('pick a cell: row |1:10|, column |1:10|. Input format: 3,6\n')
        row = cell.split(',')[0]
        column = cell.split(',')[1]
        print(f'You chose {row} x {column}')

    def structure(self):
        print(f'Field size: {self.mode} x {self.mode}')
        block = ['_' for i in range(self.mode)]

        print('\t ', ' '.join(map(str, np.arange(1, self.mode+1))) )
        print('\t', ' '.join(block))
        
        for i in range(self.mode):
            print(i+1, '\t|', ' '.join(block))

    def mines(self):
        print('planting the mines')

    def time_track(self):
        print(f'time passed: {time.perf_counter()}')
```

```python
if __name__ == '__main__':
    game = MineSweeper('simple')
    game.structure()
    # game.game_start()
```
